# Data Wrangling

```python
np.random.seed(1)
df = DataFrame({'City': ['Mumbai','Mumbai','Mumbai','Pune','Pune','Pune','Hyderabad','Hyderabad','Chennai','Chennai'],
                  'Gender': ['Male','Male','Female','Female','Female','Male','Female','Male','Male','Female'],
                 'Marks': np.random.randint(1,100,10)})
df
```

```python
# Data wrangling function such as pivot, pivot table,crosstab would convert unique values of the columns passed in index
# and columns argument as labels of index and columns 
```

```python
df.pivot_table(values='Marks',index='City',columns='Gender')
```

```python
df.pivot_table(values = 'Marks',index='City',columns='Gender',aggfunc=min)
```

```python
df.pivot_table(values = 'Marks',index='City',columns = 'Gender', aggfunc = max)
```

```python
df.pivot_table(values = 'Marks',index='City',columns='Gender',aggfunc=sum)
```

```python
df.pivot_table(values = 'Marks',index = 'City',columns='Gender',aggfunc='count')
```

# CrossTab

```python
pd.crosstab(df['City'],df['Gender'])
```

```python
pd.crosstab(df['City'],df['Gender'],margins = True)
```

# Merges

Inner
Outer
Left
Right

```python
df1 = DataFrame({'Key':list('ABC'),'Data 1':[1,2,3]})
df2 = DataFrame({'Key':list('XBZ'),'Data 2':[100,200,30]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,how='inner')
```

```python
pd.merge(df1,df2,how='outer')
```

```python
pd.merge(df1,df2,how='left')
```

```python
pd.merge(df1,df2,how='right')
```

```python
pd.merge(df1,df2,how='right',on='Key')
```

```python
df1 = DataFrame({'Key 1':list('ABC'),'Data 1':[1,2,3]})
df2 = DataFrame({'Key 2': list('XBZ'),'Data 2':[100,200,30]})
```

```python
pd.merge(df1,df2,left_on='Key 1',right_on='Key 2')
```
