# Donut

```python
circle = plt.Circle((0,0),radius=0.5,color='white')
plt.pie(data,labels=particular)
donut= plt.gcf()                      # get current figure(gcf)
donut.gca().add_artist(circle)        # get current axis(gca)
plt.show()
```

```python
particular = ['Rent', 'Fuel','Food','Internet','Miss']
data = [10000,3000,5000,1000,2000]
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
t = df.species.value_counts()
t
```

```python
circle = plt.Circle((0,0), radius = 0.5, color = 'white')
plt.pie(t, labels = ['setosa', 'versicolor', 'virginica'])
donut = plt.gcf()
plt.gca().add_artist(circle)
plt.show()
```

```python
circle = plt.Circle((0,0), radius = 0.5, color = 'white')
plt.pie(data, labels = particular)
donut = plt.gcf()
donut.gca().add_artist(circle)
plt.show()
```
